```python
num=['Age', 'Gender', 'Region', 'Policy_Type', 'Premium_Amount',
       'Hospital_Stay_Days', 'Claim_Amount', 'Chronic_Condition',
       'Claim_Processing_Time_Days', 'Claim_Approved']
def univariate_analysis(n):
    plt.figure(figsize=(25, 30))
    for i, col in enumerate(n, 1):
        plt.subplot(5,2,i)
        plt.title(f"Count of {col}")
        sns.boxplot(x=df[col])
    plt.show()

univariate_analysis(num)
```

```python
def cap_outliers_iqr(df,column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    capped_column_name = f'{column_name}_Capped'
    df[capped_column_name] = df[column_name]
    df[capped_column_name] = np.where(
        df[capped_column_name] > upper_bound,
        upper_bound,
        df[capped_column_name]
    )
    print(f'{column_name}')
    print(f'upper bound calculated:{upper_bound:,.2f}')
    print(f'outliers capped in column:{capped_column_name}')
    return df

df = cap_outliers_iqr(df,'Premium_Amount')
```

```python
df = cap_outliers_iqr(df,'Claim_Amount')
```
