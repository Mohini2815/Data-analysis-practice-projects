```python
def cap_outliers_iqr(df,column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    capped_column_name = f'{column_name}_Capped'
    df[capped_column_name] = df[column_name]
    df[capped_column_name] = np.where(
        df[capped_column_name] > upper_bound,
        upper_bound,
        df[capped_column_name]
    )
    print(f'{column_name}')
    print(f'upper bound calculated:{upper_bound:,.2f}')
    print(f'outliers capped in column:{capped_column_name}')
    return df

df = cap_outliers_iqr(df,'Premium_Amount')
```

```python
stay_days_array = df['Hospital_Stay_Days'].values
stay_days_array
```

1. np.mean() - calculates the average value.
2. np.median() - calculates the middle value.
3. np.std() - calculates the standard deviation(spread).

```python
average_stay = np.mean(stay_days_array)
median_stay = np.median(stay_days_array)
std_stay = np.std(stay_days_array)

print(f'Average hospital stay days: {average_stay:.2f}')
print(f'Median hospital stay days: {median_stay}')
print(f'standard deviation hospital stay days: {std_stay:.2f}')
```

```python
mean_age = np.mean(df['Age'])
std_age = np.std(df['Age'])
print(f'mean age: {np.round(mean_age,2)}')
print(f'standard deviation of age: {np.round(std_age,2)}')
```

1. np.min() - finds smallest value.
