```python
# Imputation Strategy: Group-wise Median ---
# Calculate the median Volume_Traded_USD for each Coin_Name
coin_median_volume = df.groupby('Coin_Name')['Volume_Traded_USD'].median()
df['Volume_Traded_USD'] = df['Volume_Traded_USD']

# Fill missing values in 'Volume_Traded_USD' using the corresponding coin's median
df['Volume_Traded_USD'] = df.apply(
    lambda row: coin_median_volume[row['Coin_Name']] if pd.isna(row['Volume_Traded_USD']) else row['Volume_Traded_USD'],
    axis=1
)

# --- Verification (and fallback for edge cases) ---
# If any values are still missing, fill them with the overall median
remaining_na = df['Volume_Traded_USD'].isnull().sum()
if remaining_na > 0:
    overall_median_volume = df['Volume_Traded_USD'].median()
    df['Volume_Traded_USD'].fillna(overall_median_volume, inplace=True)
```

```python
# calculate median 
median_count = df['Transactions_Count'].median()
df['Transactions_Count'].fillna(median_count,inplace=True)
```
