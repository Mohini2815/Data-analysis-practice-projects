```python
# risk category in loan amount by city
risk_by_city_pivot = pd.crosstab(
    df['City'],
    df['Risk_Category'],
    margins = True, # add all rows or columns for totals
    margins_name = 'Loan_Amount'
)

# calculate percentage distribution 
# will show the percentage breakdown of risk for each city
risk_by_city_pct = (
    pd.crosstab(df['City'],df['Risk_Category'],normalize='index') * 100 
).round(2)
# sort by the high risk percentage 
# if high is a category to easily identify the riskest cities 
if 'High' in risk_by_city_pct.columns:
    risk_by_city_pct_sorted = risk_by_city_pct.sort_values(by='High',
                                                          ascending = False) # fallback sort
else:
    # if high is not available , sort by a common risk indicator like 'medium' or 'a specific risk category'
    risk_by_city_pct_sorted = risk_by_city_pct.sort_values(by = risk_by_city_pct.columns[0],ascending=False)

print(risk_by_city_pivot)
```

```python
# Approval status by loan amount and income 
# calculates averages by approval status
status_summary = df.groupby('Approval_Status')[['Loan_Amount','Income']].mean().round(2)
print(status_summary)
```

```python
fig,axes = plt.subplots(1,2,figsize=(14,6))
df.boxplot(column='Loan_Amount',by= 'Approval_Status',ax=axes[0])
axes[0].set_title('loan amount distributions')
axes[0].set_ylabel('loan amount')
axes[0].set_xlabel('approval status')
axes[0].grid(False)

df.boxplot(column='Income',by='Approval_Status',ax=axes[1])
axes[1].set_title('income distributions')
axes[1].set_ylabel('income')
axes[1].set_xlabel('approval status')
axes[1].grid(False)

plt.suptitle('comparison of loan amount and income for approved vs rejected loans',y=1.02)
plt.tight_layout()
plt.savefig('approval status by')
```
