```python
# risk category in loan amount by city
risk_by_city_pivot = pd.crosstab(
    df['City'],
    df['Risk_Category'],
    margins = True, # add all rows or columns for totals
    margins_name = 'Loan_Amount'
)

# calculate percentage distribution 
# will show the percentage breakdown of risk for each city
risk_by_city_pct = (
    pd.crosstab(df['City'],df['Risk_Category'],normalize='index') * 100 
).round(2)
# sort by the high risk percentage 
# if high is a category to easily identify the riskest cities 
if 'High' in risk_by_city_pct.columns:
    risk_by_city_pct_sorted = risk_by_city_pct.sort_values(by='High',
                                                          ascending = False) # fallback sort
else:
    # if high is not available , sort by a common risk indicator like 'medium' or 'a specific risk category'
    risk_by_city_pct_sorted = risk_by_city_pct.sort_values(by = risk_by_city_pct.columns[0],ascending=False)

print(risk_by_city_pivot)
```

2. Income distributions

```python
# how does debt_to_income_ratio interact with credit_score and income
# calculate correlation matrix
corr_matrix = df[['Debt_To_Income_Ratio','Credit_Score','Income']].corr()
print('Correlatio matrix = ')
print(corr_matrix)
```

```python
# visualize correlation matrix (heatmap)
plt.figure(figsize=(10,2))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f',linewidth=.5)
plt.title('Correlation between DTI,credit score and income is ')
plt.show()
```

1. DTI and income
- Extremely weak negative correlation(-0.01).
- There is almost no linear relationship between a borrower's income and their debt to income ratio.

2. DTI and Credit score
- Here is also extremely negative weak correlation(-0.01).

3. Credit score and income
- Weak positive correlation(0.14).
- There is a slight tendency for people with higher incomes to have higher credit score.     

```python
# create bins for credit score(low.medium.high)
score_bins = [500,650,750,900]
score_labels = ['Low','Medium','High']
df['credit_score_group'] = pd.cut(df['Credit_Score'],bins = score_bins,labels = score_labels,right=False)

# group by the created bins and find the average DTI
dti_by_score = df.groupby('credit_score_group')['Debt_To_Income_Ratio'].mean().reset_index()

print('Average dti by credit score group=')
print(dti_by_score)
```

```python
plt.figure(figsize=(10,2))
sns.barplot(x='credit_score_group',y='Debt_To_Income_Ratio',data=dti_by_score)
plt.title('Average DTI by credit score')
plt.ylabel('Average DTI')
plt.xlabel('credit score tier')
plt.show()
```

```python
TENURE_COL = 'Loan_Tenure_Months'
STATUS_COL = 'Approval_Status'



# --- 2. Loan Tenure Violin Plot by Approval Status ---
sns.set_theme(style="whitegrid")

# Create a figure for the single plot
plt.figure(figsize=(9, 7))

# Generate the violin plot for Loan Tenure, categorized by Approval Status
# 'inner="quartile"' draws lines for the median (white dot) and the 25th/75th quartiles.
# 'split=True' can also be used if there are only two categories, creating a half-violin plot.
sns.violinplot(
    x=STATUS_COL,
    y=TENURE_COL,
    data=df,
    palette={"Approved": "#2ECC71", "Rejected": "#E74C3C"}, # Custom colors
    inner="quartile",
    cut=0,
    linewidth=1.5
)


plt.title('Loan Tenure Distribution by Approval Status (Violin Plot)', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Loan Approval Status', fontsize=13)
plt.ylabel('Loan Tenure (Months)', fontsize=13)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

print("\n--- Plot Description and Insights ---")
print("This violin plot compares the probability distribution of Loan Tenure between Approved and Rejected applications.")
print("The width of the shape at any given point indicates the density (frequency) of loans at that specific tenure value.")
print("The inner lines represent the quartiles (25th, 50th/Median, and 75th percentiles), allowing you to quickly compare central tendencies.")
print(f"Total valid records analyzed: {len(df)}")
```

```python
TENURE_COL = 'Loan_Tenure_Months'
AGE_COL = 'Age'

# Calculate the average loan tenure for each unique age
# We group by Age and then calculate the mean Loan_Tenure_Months
age_tenure_avg = df.groupby(AGE_COL)[TENURE_COL].mean().reset_index()

# Sort the data by Age to ensure the line plot connects points correctly
age_tenure_avg.sort_values(AGE_COL, inplace=True)


# --- 3. Plotting the Line Chart ---
sns.set_theme(style="whitegrid")

plt.figure(figsize=(12, 6))

# Create the line plot
sns.lineplot(
    x=AGE_COL,
    y=TENURE_COL,
    data=age_tenure_avg,
    marker='o', # Add circles for each data point
    color="#FF5733", # A distinct color
    linewidth=2
)

plt.title('Trend of Average Loan Tenure by Applicant Age', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Applicant Age', fontsize=13)
plt.ylabel('Average Loan Tenure (Months)', fontsize=13)

# Ensure x-axis ticks are readable, showing every 5th year for example
min_age = age_tenure_avg[AGE_COL].min()
max_age = age_tenure_avg[AGE_COL].max()
# Calculate ticks in steps of 5, ensuring min/max are included
ticks = [age for age in range(int(min_age), int(max_age) + 1, 5)]
if ticks[-1] != max_age:
    ticks.append(max_age)

plt.xticks(ticks=ticks, rotation=45, ha='right')
plt.grid(axis='both', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

print("\n--- Line Plot Interpretation ---")
print("This line plot shows the relationship between an applicant's age and the average loan tenure they receive.")
print("A clear trend (upward or downward line) indicates a correlation between age and the duration of the loan.")
print(f"Number of unique ages analyzed: {len(age_tenure_avg)}")
```
